## Correlation Analysis

```python
# How much does each factor correlate with happiness?

df.corr(numeric_only=True)['Score'].sort_values(ascending=False).round(2)
```

This means money alone doesn't make you happy — you also need health and people who care about you. 

```python
# Countries with high GDP but below average happiness

avg_happiness = df['Score'].mean()
avg_wealthy= df['GDP per capita'].mean()

wealthy_unhappy = df[
                    (df['GDP per capita'] > avg_wealthy) &
                    (df['Score'] < avg_happiness)
                    ][['Country or region', 'Score', 'GDP per capita', 'Social support']].sort_values('Score',ascending=  False)

print(wealthy_unhappy)
```

Middle East & North African countries have high GDP because of its high resource and geography.but governance issues, inequality, and political instability drag happiness down.

```python
# Countries with low GDP but above average happiness

poor_happy = df[
                (df['GDP per capita'] < avg_wealthy) &
                (df['Score']>avg_happiness)
                 ][['Country or region', 'Score', 'GDP per capita', 'Social support']].sort_values('Score',ascending=  False)

print(poor_happy )
```

LATIN American countries shows consistently score high on happiness despite lower wealth. Their Social support scores — consistently above 1.2. Strong families, tight communities, rich culture. People show up for each other.
    

```python
# Compare average social support — Latin America vs wealthy unhappy countries

latin = ['Guatemala', 'El Salvador', 'Nicaragua', 'Honduras', 'Bolivia', 'Jamaica']
rich_unhappy_countries = ['Greece', 'Turkey', 'Malaysia', 'Belarus']

print("Latin America avg social support:")
print(df[df['Country or region'].isin(latin)]['Social support'].mean().round(3))

print("\nRich unhappy countries avg social support:")
print(df[df['Country or region'].isin(rich_unhappy_countries)]['Social support'].mean().round(3))
```

```python
# Countries with high health expectancy but above average happiness

avg_happiness = df['Score'].mean()
avg_healthy= df['Healthy life expectancy'].mean()

healthy_unhappy = df[
                    (df['Healthy life expectancy'] > avg_healthy) &
                    (df['Score'] < avg_happiness)
                    ][['Country or region', 'Score', 'Healthy life expectancy', 'Perceptions of corruption','Freedom to make life choices']].sort_values('Score',ascending=  False)

print(healthy_unhappy)
```

Eastern European countries have good healthcare systems , but low happiness due to corruption and low freedom scores. Also,some Asian countries — long life expectancy culturally, but less personal freedom.

```python
# Countries with low health expectancy but above average happiness

avg_happiness = df['Score'].mean()
avg_healthy= df['Healthy life expectancy'].mean()

low_healthy_happy = df[
                    (df['Healthy life expectancy'] < avg_healthy) &
                    (df['Score'] > avg_happiness)
                    ][['Country or region', 'Score', 'Healthy life expectancy', 'Perceptions of corruption','Freedom to make life choices','Social support']].sort_values('Score',ascending=  False)

print(low_healthy_happy)
```

```python
df[df['Country or region'].isin(['Trinidad & Tobago', 'Bolivia', 'Pakistan'])]
```

Pakistan has the worst health score in this list and the lowest freedom score — yet still above average happiness. Strong religious faith and tight family structures. Pakistani culture has extremely deep family bonds — people rarely face hardship alone.

Trinidad & Tobago has the highest happiness in this group despite poor health. But look at that corruption score — 0.016 — almost zero perceived corruption. People trust their government and each other. That trust alone boosts happiness significantly.

## Conclusion

- GDP correlates 0.79 - Wealth matters most numerically
- Greece, Turkey unhappy despite wealth - Money without freedom/community fails
- Guatemala, Nicaragua happy despite poverty - Community beats wealth
- Pakistan, Philippines happy despite poor health - Culture and faith fill the gap

Happiness isn't one thing ,it's a combination of wealth, health, freedom AND community. No single factor guarantees it

```python
# Drop duplicates if any
df = df.drop_duplicates()

# Fill missing numbers with column median (safer than mean — remember outliers!)
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Reset index after cleaning
df = df.reset_index(drop=True)

print(f"Clean dataset shape: {df.shape}")
print(df.isnull().sum())
```

```python
df.to_csv('happiness_clean.csv', index=False)
print("Saved!")
```
